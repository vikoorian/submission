## Visualization & Explanatory Analysis

```python
# Mengubah kolom 'dteday' menjadi tipe data datetime
bike_hour_df['dteday'] = pd.to_datetime(bike_hour_df['dteday'])
```

```python
# Ekstraksi bulan dari kolom 'dteday'
bike_hour_df['month'] = bike_hour_df['dteday'].dt.month
```

```python
# Menghitung jumlah pengguna sepeda berdasarkan bulan
monthly_usage = bike_hour_df.groupby('month')['cnt'].sum().reset_index()
```

### Pertanyaan 1:

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='month', y='cnt', data=monthly_usage)
max_value = monthly_usage['cnt'].max()  # Mendapatkan jumlah pengguna sepeda tertinggi
max_index = monthly_usage['cnt'].idxmax()  # Mendapatkan indeks bulan dengan jumlah pengguna sepeda tertinggi

# Mengubah warna batang tertinggi menjadi merah
ax.patches[max_index].set_facecolor('red')
ax.set_axisbelow(True) 

# Mengubah warna batang lain menjadi abu-abu
for i, patch in enumerate(ax.patches):
    if i != max_index:
        patch.set_facecolor('gray')

plt.xlabel('Bulan')
plt.ylabel('Jumlah Pengguna')
plt.title('Jumlah Pengguna Sepeda per Bulan')
bulan = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
plt.xticks(ticks=range(12), labels=bulan, rotation=45)
plt.grid(which='both', axis='y', linestyle='-', alpha=0.8, zorder=0)  # Menambah garis bantu pada sumbu y
ax.set_zorder(1)  # Memastikan diagram batang ada di atas garis bantu
plt.tight_layout()
plt.savefig('jumlah_pengguna_sepeda_per_bulan.png')  # Menyimpan visualisasi
plt.show()
```

### Pertanyaan 2:

```python
plt.figure(figsize=(10,6))
sns.heatmap(bike_hour_df[['temp', 'hum', 'windspeed', 'cnt']].corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi antara Variabel Cuaca dan Jumlah Pengguna Sepeda')
plt.show()
```

## Conclusion

- Conclution pertanyaan 1

    Dari analisis pola penggunaan sepeda berdasarkan bulan, dapat dilihat bahwa penggunaan sepeda cenderung lebih tinggi di bulan-bulan dengan cuaca yang lebih hangat, seperti bulan-bulan musim panas. Bulan-bulan dengan cuaca dingin cenderung memiliki jumlah pengguna sepeda yang lebih rendah. lebih tinggi di bulan-bulan dengan cuaca yang lebih hangat.
- Conclution pertanyaan 2

    Hasil analisis korelasi antara variabel cuaca (suhu, kelembaban, kecepatan angin) dan jumlah pengguna sepeda menunjukkan bahwa suhu memiliki korelasi positif yang signifikan dengan penggunaan sepeda. Artinya, semakin tinggi suhu, semakin tinggi juga jumlah pengguna sepeda. Di sisi lain, kelembaban dan kecepatan angin memiliki korelasi negatif yang lebih rendah dengan penggunaan sepeda, meskipun tidak sekuat korelasi positif suhu.

    Dari kedua temuan ini, dapat disimpulkan bahwa faktor cuaca seperti suhu memengaruhi penggunaan sepeda secara signifikan, sedangkan faktor lain seperti kelembaban dan kecepatan angin memiliki pengaruh yang lebih kecil. Oleh karena itu, pemahaman tentang faktor-faktor cuaca ini dapat membantu dalam perencanaan dan pengelolaan layanan penyewaan sepeda, terutama dalam menentukan waktu dan lokasi promosi serta alokasi sumber daya.
